```python
import boto3
import pandas as pd
import os
```

```python
s3_resource = boto3.resource("s3")
s3_client = s3_resource.meta.client

objects = s3_client.list_objects(Bucket="wri-public-data",
                                Prefix="resourcewatch/",
                                 # Use Delimiter to only see the folders in resourcewatch/raster/
                                 # To see the actual items, need to remove the Delimiter keyword
                                #Delimiter="/"
                                )

all_objects = [obj for obj in objects["Contents"] if obj["Key"][-1] != "/"]
all_objects = pd.DataFrame(all_objects)
all_objects
```

```python
# List comes from the July tracking sheet, filter Planet Pulse/RT column for [RT/pulse/explore/both/????] 
list_of_nrt_datasets = ['bio.005','cit.003','cit.018','cit.026',
'cli.002','cli.003','cli.005','cli.012','cli.013','cli.021','cli.023',
'cli.024','cli.025','cli.027','cli.034','cli.035','cli.036','cli.037',
'dis.001','dis.003','dis.005','dis.007','dis.009','dis.012','ene.008',
'foo.003','foo.021','foo.023','foo.024','for.002','for.003','for.006',
'for.010','for.015','soc.016','soc.017','soc.018','soc.038','soc.043',
'soc.047','wat.025']

list_of_nrt_datasets = [rw_id.replace(".", "_") for rw_id in list_of_nrt_datasets]
list_of_nrt_datasets
```

```python
#https://leemendelowitz.github.io/blog/any-all-in-python.html

def filter_for_nrt(key):
    return(any(rw_id in key for rw_id in list_of_nrt_datasets))
all_objects["nrt"] = list(map(filter_for_nrt, all_objects["Key"]))
nrt_objects = all_objects[all_objects["nrt"]]
print(nrt_objects.shape)
nrt_objects
```
